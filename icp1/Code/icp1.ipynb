```
myDF = pd.read_csv('drive/My Drive/Data.csv')
mapping = {'M': 0, 'B': 1 }
myDF = myDF.replace({'diagnosis': mapping})
myDF
```

```
corr=myDF.corr()
corr
```

```
import seaborn as sns
from matplotlib import pyplot as plt
plt.figure(figsize=(40,40))
sns.heatmap(corr, annot=True,cmap='BuGn')
plt.show()
```
